#### A1 [05 marks]
#### Display the customer surname (and the value) that has the highest repeat transactions or repeat purchase. (Tip: Use count or sum statistic)

```python
# A1

import matplotlib.pyplot as plt
import seaborn as sns

surname_counts = my_df['surname'].value_counts()
top_surnames = surname_counts.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_surnames.values, y=top_surnames.index, palette="viridis")
plt.title("Top 10 Most Frequent Customer Surnames")
plt.xlabel("Frequency")
plt.ylabel("Surname")
plt.show()

most_frequent_surname = surname_counts.idxmax()
frequency_of_most_frequent_surname = surname_counts.max()

most_frequent_surname, frequency_of_most_frequent_surname
```

#### A2 [05 marks]
#### Display the country (and the value) that has: (1) the highest number of membership, (2) the lowest number of membership. (Tip: Use member = 1 | Use count or sum statistic)

```python
# A2

membership_counts = my_df[my_df['member'] == 1]['country'].value_counts()

# Finding the country with the highest and lowest number of memberships
country_with_highest_membership = membership_counts.idxmax()
highest_membership_count = membership_counts.max()

country_with_lowest_membership = membership_counts.idxmin()
lowest_membership_count = membership_counts.min()

# Filtering the data to include only the highest and lowest countries for memberships
filtered_membership_counts = membership_counts.loc[[country_with_highest_membership, country_with_lowest_membership]]

# Plotting the membership counts for the countries with the highest and lowest memberships
plt.figure(figsize=(6, 4))
sns.barplot(x=filtered_membership_counts.index, y=filtered_membership_counts.values, palette="coolwarm")
plt.title("Countries with the Highest and Lowest Number of Memberships")
plt.xlabel("Country")
plt.ylabel("Number of Memberships")
plt.show()

country_with_highest_membership, highest_membership_count, country_with_lowest_membership, lowest_membership_count
```

#### A3 [05 marks]
#### Display the gender (and the value) that is: (1) the youngest, (2) the oldest. (Tip: Use median statistic)

```python
# A3

# Calculating the median age for each gender
median_ages = my_df.groupby('gender')['age'].median()

# Finding the gender with the youngest and oldest median age
youngest_gender = median_ages.idxmin()
youngest_median_age = median_ages.min()

oldest_gender = median_ages.idxmax()
oldest_median_age = median_ages.max()

# Plotting the median ages by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=median_ages.index, y=median_ages.values, palette="pastel")
plt.title("Median Age by Gender")
plt.xlabel("Gender")
plt.ylabel("Median Age")
plt.show()

youngest_gender, youngest_median_age, oldest_gender, oldest_median_age
```

#### A4 [05 marks]
#### Display the country (and the value) that has: (1) the richest customer, (2) the poorest customers. (Tip: Use salary | Use mean statistic)

```python
# A4

# Calculating the mean salary for each country
mean_salaries = my_df.groupby('country')['salary'].mean()

# Finding the country with the highest and lowest mean salary
richest_country = mean_salaries.idxmax()
richest_mean_salary = mean_salaries.max()

poorest_country = mean_salaries.idxmin()
poorest_mean_salary = mean_salaries.min()

# Filtering the data to include only the countries with the highest and lowest mean salaries
filtered_mean_salaries = mean_salaries.loc[[richest_country, poorest_country]]

# Plotting the mean salaries for the countries with the highest and lowest mean salaries
plt.figure(figsize=(6, 4))
sns.barplot(x=filtered_mean_salaries.index, y=filtered_mean_salaries.values, palette="cubehelix")
plt.title("Countries with the Highest and Lowest Mean Salaries")
plt.xlabel("Country")
plt.ylabel("Mean Salary")
plt.show()

richest_country, richest_mean_salary, poorest_country, poorest_mean_salary
```

#### A5 [05 marks]
#### Display the gender (and the value) that has: (1) the best credit score, (2) the worst credit score. (Tip: Use mean statictic)

```python
# A5

# Calculating the mean credit score for each gender
mean_credit_scores = my_df.groupby('gender')['credit_score'].mean()

# Finding the gender with the best (highest) and worst (lowest) mean credit score
best_credit_gender = mean_credit_scores.idxmax()
best_mean_credit_score = mean_credit_scores.max()

worst_credit_gender = mean_credit_scores.idxmin()
worst_mean_credit_score = mean_credit_scores.min()

# Plotting the mean credit scores by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=mean_credit_scores.index, y=mean_credit_scores.values, palette="husl")
plt.title("Mean Credit Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Mean Credit Score")
plt.show()

best_credit_gender, best_mean_credit_score, worst_credit_gender, worst_mean_credit_score
```

#### A6 [05 marks]
#### Display the country (and the value) with: (1) highest variation in salary (2) lowest variation in salary. (Tip: Use std. dev. statistic)

```python
# A6

# Calculating the standard deviation of salary for each country
salary_std_dev = my_df.groupby('country')['salary'].std()

# Finding the country with the highest and lowest standard deviation in salary
highest_salary_variation_country = salary_std_dev.idxmax()
highest_salary_variation = salary_std_dev.max()

lowest_salary_variation_country = salary_std_dev.idxmin()
lowest_salary_variation = salary_std_dev.min()

# Filtering the data to include only the countries with the highest and lowest standard deviation in salary
filtered_salary_std_dev = salary_std_dev.loc[[highest_salary_variation_country, lowest_salary_variation_country]]

# Plotting the standard deviation of salaries for the countries with the highest and lowest variation
plt.figure(figsize=(6, 4))
sns.barplot(x=filtered_salary_std_dev.index, y=filtered_salary_std_dev.values, palette="Paired")
plt.title("Countries with the Highest and Lowest Variation in Salary (Standard Deviation)")
plt.xlabel("Country")
plt.ylabel("Salary Standard Deviation")
plt.show()

highest_salary_variation_country, highest_salary_variation, lowest_salary_variation_country, lowest_salary_variation
```

#### A7 [05 marks]
#### Display the gender statistics with respect to country. (Tip: Use count or sum statistic)
##### 1. Country with highest number of male customer 2. Country with lowest number of male customer
##### 3. Country with highest number of female customer 4. Country with lowest number of female customer

```python
# A7

# Counting the number of male and female customers by country
gender_counts_by_country = my_df.groupby(['country', 'gender']).size().unstack()

# Identifying the countries with the highest and lowest number of male and female customers
highest_male_country = gender_counts_by_country['Male'].idxmax()
highest_male_count = gender_counts_by_country['Male'].max()

lowest_male_country = gender_counts_by_country['Male'].idxmin()
lowest_male_count = gender_counts_by_country['Male'].min()

highest_female_country = gender_counts_by_country['Female'].idxmax()
highest_female_count = gender_counts_by_country['Female'].max()

lowest_female_country = gender_counts_by_country['Female'].idxmin()
lowest_female_count = gender_counts_by_country['Female'].min()

# Plotting the gender statistics with respect to country
plt.figure(figsize=(10, 6))
gender_counts_by_country.plot(kind='bar', stacked=False, color=["red", "orange"])
plt.title("Gender Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Number of Customers")
plt.legend(title='Gender')
plt.show()

{
    "highest_male": (highest_male_country, highest_male_count),
    "lowest_male": (lowest_male_country, lowest_male_count),
    "highest_female": (highest_female_country, highest_female_count),
    "lowest_female": (lowest_female_country, lowest_female_count)
}
```

####  A9 [05 marks]
#### Display the combination of country & gender (and the value) that has: (1) the best credit score, (2) the worst credit score. (Tip: Use mean statictic)

```python
# A9

# Calculating the mean credit score for each combination of country and gender
mean_credit_scores_by_country_gender = my_df.groupby(['country', 'gender'])['credit_score'].mean()

# Finding the combination with the best (highest) and worst (lowest) mean credit score
best_credit_combo = mean_credit_scores_by_country_gender.idxmax()
best_credit_score = mean_credit_scores_by_country_gender.max()

worst_credit_combo = mean_credit_scores_by_country_gender.idxmin()
worst_credit_score = mean_credit_scores_by_country_gender.min()

# Filtering the data for the highest and lowest combinations
filtered_credit_scores = mean_credit_scores_by_country_gender.loc[[best_credit_combo, worst_credit_combo]]

# Plotting the mean credit scores for the combinations with the highest and lowest scores
plt.figure(figsize=(6, 4))
sns.barplot(x=filtered_credit_scores.index.map(lambda x: f"{x[0]}, {x[1]}"), y=filtered_credit_scores.values, palette="hls")
plt.title("Best and Worst Credit Scores by Country and Gender")
plt.xlabel("Country, Gender")
plt.ylabel("Mean Credit Score")
plt.show()

best_credit_combo, best_credit_score, worst_credit_combo, worst_credit_score
```

#### A10 [05 marks]
#### Display the top 3 & bottom 3 credit scores in the: (1) members catergory, (2) non-members category. (Tip: Use n-largest / n-smallest statictic)

```python
# A10

# Finding the top 3 and bottom 3 credit scores in the members category
top_3_member_credit_scores = my_df[my_df['member'] == 1]['credit_score'].nlargest(3)
bottom_3_member_credit_scores = my_df[my_df['member'] == 1]['credit_score'].nsmallest(3)

# Finding the top 3 and bottom 3 credit scores in the non-members category
top_3_non_member_credit_scores = my_df[my_df['member'] == 0]['credit_score'].nlargest(3)
bottom_3_non_member_credit_scores = my_df[my_df['member'] == 0]['credit_score'].nsmallest(3)

# Combining these into a single dataframe for plotting
credit_scores_combined = pd.DataFrame({
    'Top 3 Members': top_3_member_credit_scores.values,
    'Bottom 3 Members': bottom_3_member_credit_scores.values,
    'Top 3 Non-Members': top_3_non_member_credit_scores.values,
    'Bottom 3 Non-Members': bottom_3_non_member_credit_scores.values
})

# Plotting the top 3 and bottom 3 credit scores for members and non-members
plt.figure(figsize=(10, 6))
credit_scores_combined.plot(kind='bar', color=["lightblue", "darkblue", "tan", "red"])
plt.title("Top 3 and Bottom 3 Credit Scores in Members and Non-Members Categories")
plt.xlabel("Rank")
plt.ylabel("Credit Score")
plt.xticks(ticks=[0, 1, 2], labels=['1st', '2nd', '3rd'], rotation=0)
plt.legend(title='Category')
plt.show()

credit_scores_combined
```

#### A11 [05 marks]

#### A12 [05 marks]
#### Display the top 2 & bottom 2 age groups in the (1) members category, (2) non-members category. (Tip: Use n-largest / n-smallest statistic)

```python
# A12

bins = [18, 30, 40, 50, 60, 70, 80]
labels = ['18-30', '31-40', '41-50', '51-60', '61-70', '71-80']
my_df['age_group'] = pd.cut(my_df['age'], bins=bins, labels=labels, right=False)

# Counting the number of customers in each age group for members and non-members
age_group_counts_members = my_df[my_df['member'] == 1]['age_group'].value_counts()
age_group_counts_non_members = my_df[my_df['member'] == 0]['age_group'].value_counts()

# Creating a DataFrame to combine the values for plotting
combined_age_groups = pd.DataFrame({
    "Members": age_group_counts_members,
    "Non-Members": age_group_counts_non_members
})

# Extracting the top 2 and bottom 2 age groups for members and non-members
top_2_members = combined_age_groups['Members'].nlargest(2)
bottom_2_members = combined_age_groups['Members'].nsmallest(2)
top_2_non_members = combined_age_groups['Non-Members'].nlargest(2)
bottom_2_non_members = combined_age_groups['Non-Members'].nsmallest(2)

# Merging top 2 and bottom 2 age groups into a single DataFrame for plotting
# Ensuring all relevant age groups are included in the index
all_age_groups = top_2_members.index.union(bottom_2_members.index).union(top_2_non_members.index).union(bottom_2_non_members.index)
merged_age_groups = pd.DataFrame(index=all_age_groups)

# Adding the top 2 and bottom 2 age groups to the DataFrame
merged_age_groups['Top 2 Members'] = top_2_members.reindex(all_age_groups)
merged_age_groups['Bottom 2 Members'] = bottom_2_members.reindex(all_age_groups)
merged_age_groups['Top 2 Non-Members'] = top_2_non_members.reindex(all_age_groups)
merged_age_groups['Bottom 2 Non-Members'] = bottom_2_non_members.reindex(all_age_groups)

# Plotting the age group counts
plt.figure(figsize=(10, 6))
merged_age_groups.plot(kind='bar', color=["pink", "orchid", "darkblue", "lightblue"])
plt.title("Top 2 & Bottom 2 Age Groups in Members and Non-Members Categories")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.legend(title='Category')
plt.show()

merged_age_groups, top_2_members, bottom_2_members, top_2_non_members, bottom_2_non_members
```
